```python
drop_columns = ['opinion_h1n1_vacc_effective', 'opinion_h1n1_risk',
       'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'agegrp',
       'raceeth4_i', 'sex_i', 'marital',
       'rent_own_r', 'employment_status', 'census_region', 'census_msa',
       'n_adult_r', 'household_children', 'n_people_r', 'employment_industry',
       'employment_occupation', 'hhs_region', 'state']
```

```python
df = df.drop(columns = drop_columns)
```

```python
test = test.drop(columns = drop_columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, val = train_test_split(df, train_size=0.70, test_size=0.30, stratify=df[target], random_state=42)
```

```python
train.shape, val.shape, test.shape
```

```python
features = train.drop(columns=[target]).columns
```

```python
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
```

```python
pipe_ord = make_pipeline( 
    SimpleImputer(), 
    RandomForestClassifier(n_estimators=500, random_state=42, max_depth=15, oob_score=True, n_jobs=-1, criterion="gini", min_samples_split=5, max_features=6)
)

pipe_ord.fit(X_train, y_train)
print('검증 정확도', pipe_ord.score(X_val, y_val))
```

```python
pipe_ord.named_steps['randomforestclassifier'].oob_score_
```

```python
y_pred_test = pipe_ord.predict(X_test)
y_pred_test = pd.Series(y_pred_test)
y_pred_test.value_counts()
```

```python
id = pd.Series(range(len(y_pred_test)))
y_pred_test = pd.Series(y_pred_test)
submission = pd.concat([id, y_pred_test], axis=1)
submission.rename(columns={0:"id", 1:target}, inplace=True)

# Display the submission data information
print(submission.shape)
print(submission.value_counts(target))

# Export the dataset to a *csv* file.
submission.to_csv("./submission.csv",index=False)
```

```python
X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
from sklearn.model_selection import GridSearchCV

params = { 'n_estimators' : [10, 100],
           'max_depth' : [6, 8, 10, 12],
           'min_samples_leaf' : [8, 12, 18],
           'min_samples_split' : [8, 16, 20]
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(X_train, y_train)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```
